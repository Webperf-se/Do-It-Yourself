```python
#import requests
from esnicheck.check import ESNICheck
from urllib.parse import urlparse
#from bs4 import BeautifulSoup

def check_esni_tls(url="https://webperf.se"):
    #r = requests.get(f"https://esnicheck.com/check?q={url}")

    #soup = BeautifulSoup(r.content, 'html.parser')
    #esni = soup.findAll('article')[1].text.replace(url, '').replace('\n', '')
    #tls = soup.findAll('article')[2].findAll("div", {"class": "message-body"})[0].text.replace(url, '').replace('\n', '')
    
    o = urlparse(url)
    host = ESNICheck(o.netloc)
    
    esni = host.has_esni()
    tls = host.has_tls13()
    dns = host.has_dns()
    
    print(f"{url};{esni};{tls};{dns}")

check_esni_tls("https://webperf.se")
```
