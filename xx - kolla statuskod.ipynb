## 2. Hämta statuskoden för URL

```python
url = 'https://1177.se/kebabibröd.html'
import urllib

points = 0
review = ''
result_dict = {}

o = urllib.parse.urlparse(url)
url = '{0}://{1}/{2}.html'.format(o.scheme, o.netloc, 's1d4-f1nns-1nt3')
headers = {'user-agent': 'Mozilla/5.0 (compatible; Googlebot/2.1; +http://www.google.com/bot.html)'}
request = requests.get(url, allow_redirects=True, headers=headers, timeout=30)
code = request.status_code
if code == 404:
    points += 2
else:
    review = review + '* Fel statuskod. Fick {0} när 404 vore korrekt.\n'.format(request.status_code)

result_dict['status_code'] = code

code
```
